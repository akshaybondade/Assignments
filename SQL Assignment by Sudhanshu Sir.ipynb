```python
#4 . Dump all the data into car table 

import mysql.connector as conn
import csv
try:
    mydb = conn.connect(host = 'localhost', database = 'cardataset2', user = 'root', passwd = '1234', use_pure = True)
    cursor = mydb.cursor()
    query = 'INSERT INTO cardataset2.car VALUES'

    with open('car.data','r') as f:
        glass_data=csv.reader(f,delimiter='\n')
        for index,item in enumerate(glass_data):
            strQ = '"'+'","'.join(item[0].split(","))+'"'

            if index != 0: query +=","
            query += "("+strQ+")"
            
            
        #cursor.execute(query)
        print(query)
        mydb.close()

except Exception as e:
    mydb.close()
    print(str(e))
        
```
